```
train = pd.read_csv('data/train_data/train_task_3_4.csv')
print(train.head(1))
print('-' * 50 + '\n')
print(train.describe())
print('-' * 50+ '\n')
print(train.info())
```

```
# calc percent of correct for each question
corr_rate = pd.DataFrame({'QuestionId':[], 'responses':[], 'correct_rate':[]})
for gid, group in train.groupby('QuestionId'):
    row = pd.DataFrame({'QuestionId':[gid], 'responses':[len(group)], 'correct_rate':[group['IsCorrect'].mean()]})
    corr_rate = pd.concat([corr_rate, row], ignore_index=True)
#
corr_rate = corr_rate.sort_values(by=['correct_rate'])
corr_rate['ranking'] = range(1, len(corr_rate)+1)
corr_rate.index = range(0, len(corr_rate))
corr_rate
```
